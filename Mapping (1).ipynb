```python
from mpl_toolkits.basemap import Basemap
from matplotlib import pyplot as plt
import numpy as np
```

```python
#create projection robin thats centered on a longitude of 0
m = Basemap(projection='robin',lon_0=0.)
```

```python
fig=plt.figure(figsize=(20,10))

#fillbackground color grey
fig.patch.set_facecolor('#e6e8ec')
#draw white map background and remove boundary
m.drawmapboundary(color='none', fill_color='white');
```

```python
#draw continents and countries

m.drawcoastlines(color='white')
m.fillcontinents(color='black', lake_color='white')
m.drawcountries(linewidth=1,color='white');
```

```python
m.drawcoastlines(color='white')
m.fillcontinents(color='black', lake_color='white')
m.drawcountries(linewidth=1,color='white');
graticule_width=20
parallels=np.arange(-90,91,graticule_width)
meridians=np.arange(-180., 181., graticule_width)
dashes=[3,5]
graticule_color='grey'
m.drawparallels(parallels,dashes=dashes,color=graticule_color)
m.drawmeridians(meridians,dashes=dashes,color=graticule_color)

plt.show()
```
